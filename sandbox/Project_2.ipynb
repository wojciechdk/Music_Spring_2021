### Prepare the data

<font color='green'>*Let's start by preeliminary cleaning of the data:*

*We will remove all rows where the city, the price and the neighbourhood is null.*</font>

```python
# Define a function that will return True if the string contains any forbidden
# characters
contains_no_forbiden_characters = f.udf(
    lambda x: not any([character
                       in list(r'0123456789%/\"[]{}') + ["'"]
                       for character in x]),
    BooleanType()
    )

# Remove all listings where city is not clearly specified
df_listings = (df_listings
  .where(
    (f.col("city").isNotNull())
    & (f.length(f.col("city")) > 2)  # Consider only cities whose name is between 2 and 20 characters long.
    & (f.length(f.col("city")) < 20)
    & (contains_no_forbiden_characters(f.col("city"))) # consider only cities that do not contain forbidden chars
    & (f.col("neighbourhood").isNotNull())
    & (f.col('price').isNotNull()))
  .withColumn('price', regexp_replace('price', ',', '')) #remove unwanted comma
  .withColumn('price', expr('substring(price, 2, length(price))').cast('float')) #Cast price to float
  .withColumn('review_scores_rating', col('review_scores_rating').cast('float')) # Cast review score to float
  )
```

```python
#Compute value
df_listing_value = (
  df_listings_rome
  .filter(f.col('review_scores_rating').isNotNull())
  .withColumn('value',
              f.col('review_scores_rating') / f.col('price'))
  .select('id',
          'price',
          'city',
          'review_scores_rating',
          'neighbourhood',
          'value')
)

df_listing_value.toPandas().head()
```

<font color='green'>*We then extract the three listings with the highest value for each neighbourghood, which is displayed in the table below*</font>

```python
from pyspark.sql.window import Window
from pyspark.sql import functions as f


window = (
  Window
  .partitionBy(df_listing_value['neighbourhood'])
  .orderBy(df_listing_value['value'].desc())
)

n = 3
n_highest_value_listing = (df_listing_value.select(col('*'), row_number().over(window).alias('rank'))
  .where(col('rank') <= n)
  .select('neighbourhood', 'value', 'id')
  .orderBy(f.asc('neighbourhood'),
           f.desc('value'))
  .toPandas())

n_highest_value_listing.head(10)
```

<font color='green'>*The table shows the neighborhoods in alphabetical order with the value and id of three most high-value listings in each neighborhood.*

*To get a better view of the listing values in each neighborhood, in order to compare the different neighborhoods, we present a table below without the listing ids with the neighborhoods as column headers.*</font>

```python
n_highest_value = (
  df_listing_value
  .select(col('*'),
          f.row_number().over(window).alias('rank'))
  .where(col('rank') <= n)
  .select('neighbourhood', 'value', 'id', 'rank')
  .orderBy(f.asc('neighbourhood'),
           f.desc('value'))
  .groupBy('rank')
  .pivot('neighbourhood')
  .max('value')
  .orderBy(f.asc('rank'))
)
n_highest_value.toPandas().head(20)
```

<font color='green'>*This table shows the three highest listing values for each neighbourhood, making it easy to compare neighborhoods. For example, Aurelio has at least three listings with a higher value than any listings in Balduina.*

Compute and visualize the popularity (i.e., number of reviews) of your city over time. Can you identify any trends? Write a paragraph describing your findings.

<font color='green'>*Let's start by creating a dataframe containing both listings and reviews for Rome:*

```python
# Compute a dataframe containing both the listings and the reviews for rome
df_reviews_rome = df_listings_rome.join(
    df_reviews.withColumnRenamed('id', 'review_id'),  # Rename column "id" from the reviews dataframe to "review_id"
    df_reviews.listing_id == df_listings_rome.id
    )
```

<font color='green'>*Now, we can create a dataframe showing the popularity of the city based on the number of reviews on a given date:*

```python
# Compute dataframe showing progress of Rome popularity per time
# Order by review date.
df_rome_popularity_over_time = (
    df_reviews_rome
    .where(f.col("date").isNotNull())  # Consider only reviews where date is specified
    .groupBy(['date'])
    .agg(
        f.countDistinct('review_id').alias('number_of_reviews')  # Count the number of reviews (Review IDs)
        )
    .orderBy(f.asc('date'))
    )

# Show the top rows
df_rome_popularity_over_time.toPandas().head(10)
```

<font color='green'>*... and visualize our findings:*

```python
# Create a pandas dataframe from spark dataframe
df_pandas_rome_popularity_over_time = df_rome_popularity_over_time.toPandas()

# Make sure that the "date" column is formatted as date
df_pandas_rome_popularity_over_time['date'] = \
 pd.to_datetime(df_pandas_rome_popularity_over_time['date'],
                format = '%Y-%m-%d').dt.date

# Create a figure and axes for the plot
figure = plt.figure(figsize=(12, 8))
axes = figure.gca()

# Plot Rome popularity over time on the created axes
df_pandas_rome_popularity_over_time.plot(
    kind='line',
    color=sns.color_palette(),
    legend=False,
    ax=axes,
    x='date',
    y=["number_of_reviews"]
 )

# Format the axes
axes.grid()
axes.set_title('Popularity of Rome over time')
axes.set_xlabel('')
axes.set_ylabel('Number of reviews')
axes.xaxis.set_major_locator(mdates.YearLocator(1, month=1, day=1))
plot = plt.setp(axes.get_xticklabels(),
                rotation=45,
                ha="right",
                va='top',
                rotation_mode="anchor")
```

<font color='green'>*It can be seen the in the figure above that the popularity of Rome has been growing steadily from 2009 until 2015, when it has reached a plateau. We can also see that the popularity is highly periodic, with minimum around January and two peaks per year: in the spring and autumn. This makes perfect sense, as these are the periods when the weather in Rome is at its best.*

```python
# Create a dataframe containg listing_id, review_id,
# and rating
df_scores = (
    df_reviews
    .select(
        f.col('id').alias('review_id'),
        'listing_id',
        'comments'
    )
    .join(
        df_listings.select(
            f.col("id").alias("listing_id"),
            f.col('review_scores_rating').alias('rating')
        ),
        on=['listing_id'],
        how='outer'
    )
    .where(
        f.col('rating').isNotNull()
        & f.col('comments').isNotNull()
    )
)

# Show the top rows
df_scores.show()
```

<font color='green'>*Now, let's tokenize the comments as follows:*

* <font color='green'>_Get only words without punctuation_ </font>
* <font color='green'>_Get only words with 3 or more letters_</font>
* <font color='green'>_Transform all words to lowercase_</font>
* <font color='green'>_Remove stopwords_</font>

```python
from pyspark.ml.feature import RegexTokenizer, StopWordsRemover, CountVectorizer

# Create a tokenizer object, that will be used to tokenize the column
# "comments" and save the result in the column "comment words".
tokenizer =  RegexTokenizer(pattern=r'\W',
                            minTokenLength=3,
                            toLowercase=True,
                            inputCol='comments',
                            outputCol='comment_words')

# Create a stopwords_remover object, that will be used to remove the
# stopwords from the column "comment words" and save the result in the
# column "comment_words_clean".
stopwords_remover = StopWordsRemover(inputCol='comment_words',
                                     outputCol='comment_words_clean')

# Tokenize and drop stopwords
df_scores_tokenized_comments = (
    stopwords_remover.transform(tokenizer.transform(df_scores))
    .drop('comment_words')
)

# Show top rows
df_scores_tokenized_comments.show(10)
```

<font color='green'>*Now, let's create a dataframe with a row for every word in every review:*

```python
# Create a dataframe with row for every word in every review
df_words = (
  df_scores_tokenized_comments
  .withColumn('words', f.explode('comment_words_clean'))
)

# Show top rows
df_words.show(10)
```

<font color='green'>*Now, we are almost ready to calculate the word score. Before we define the formula, let's look at the distribution of the review ratings:*

```python
# Create the axes and figure for the plot
figure = plt.figure(figsize=(12, 8))
axes = figure.gca()

# Plot the distribution of the review scores
(
  df_listings
 .select('review_scores_rating')
 .toPandas()
 .hist(
     column='review_scores_rating',
     bins=100,
     ax=axes
  )
)

# Format graph
axes.set_xlabel('Review score', fontdict={'size': 14})
axes.set_ylabel('Number of listings', fontdict={'size': 14})
axes.set_title('Distribution of listing ratings', fontdict={'size': 18})
```

<font color='green'>*We can see that the vast majority of the reviews have a rating of 90 or more. Therefore, we will consider a review with the score of above 90 a good review, and below 90 a bad review. To calculate the word score, we propose the following formula:*</font>


$$ \text{S} = \left( mean \left( \text{Listing rating for word} \right) - 90 \right) \cdot \log_2 \left( \text{number of word occurrences} \right) $$


<font color='green'>*The multiplication by the logarithm of word occurrences is applied to give more weight to the words that appear more often. The base of the logarithm can be used to control how much weight is given to the number of occurrences of the word with respect to the average rating of the listings in the reviews of which it appears (the lower the base, the higher the emphasis on the number of occurrences)*</font>


<font color='green'>*Now, let's use our formula to calculate the word scores and show the top rated words. We will do that using only the words that occur more than a 100 times to eliminate outliers.*</font>

```python
# Create a dataframe with word_scores
df_word_score = (
  df_words
  .groupBy('words')
  .agg(
      f.count(f.lit(1)).alias('n_occurrences'),
      f.mean('rating').alias('word_score_raw')
  )
  .withColumn(
      'word_score',
      (f.col('word_score_raw') - 90) * log2(f.col('n_occurrences'))
  )
  .where(f.col('n_occurrences') > 100)
  .orderBy(f.desc('word_score'))
)

# Show the words with the highest word score
df_word_score.show()
```

<font color='green'>*We will take a look into how the sales are distributed time-wise. Knowing when we sell more within each day, week, month and year will allow us target sales at the times where sales are lower and make sure that sales stay up when they already are.*

<font color='green'>*We will dig into which days have the greater sales, at what time of the day we are selling more, which months have the highest records and which quarters account for higher sales.*

```python
#Import functions to extract data types
from pyspark.sql.functions import year, month, dayofmonth, dayofweek, quarter

df_quarter2 = (df
  .select(
      f.col('price_EUR_api_rate').alias('price'),
      quarter('date').alias('quarter')
  )
  .where((df.date < '2019-01-01') & ((df.date >= '2016-01-01')))
  .groupBy('quarter')
  .agg(
      f.count('price').alias('order_count'),
      f.sum('price').alias('total_revenue')
  )
  .orderBy(asc('quarter'))
)

df_quarter2.toPandas()
```

```python
df_quarter2.toPandas().plot(kind = 'bar', \
                            x = 'quarter', y = 'order_count', \
                            title ='Sells per day of week', \
                            legend = False, rot =0)
```

<font color='green'> *It seems sales are concurrent to the beginning of the year. We could assume companies prefer to fill themselves with supplies for the upcoming year. Let's look at the different months:*

```python
df_month = (df
  .select(
      f.col('price_EUR_api_rate').alias('price'),
      month('date').alias('month')
  )
  .where((df.date < '2019-01-01') & ((df.date >= '2016-01-01')))
  .groupBy('month')
  .agg(
      f.count('price').alias('order_count'),
      f.sum('price').alias('total_revenue')
  )
  .orderBy(asc('month'))
)

df_month.toPandas().head(12)
```

```python
df_month.toPandas().plot(kind = 'line', x = 'month', \
                         y = 'order_count', \
                         title ='Sells per day of week', \
                         legend = False
                         color = 'black')
```

<font color='green'>*Let's dive into the distribution over the week:*

```python
df_day = (df
  .select(
      f.col('price_EUR_api_rate').alias('price'),
      dayofweek('date').alias('day_of_week')
  )
  .where((df.date < '2019-01-01') & ((df.date >= '2016-01-01')))
  .groupBy('day_of_week')
  .agg(
      f.count('price').alias('order_count'),
      f.sum('price').alias('total_revenue')
  )
  .orderBy(asc('day_of_week'))
)

df_day.toPandas()
```
